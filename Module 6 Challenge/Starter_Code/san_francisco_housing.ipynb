```python
#Use the groupby function to group the data by year. Aggregate the results by the mean of the groups.
housing_units_by_year = sfo_data.groupby('year')['housing_units'].mean()
display(housing_units_by_year)

#Use the hvplot function to plot the housing_units_by_year DataFrame as a bar chart. 
housing_units_by_year.hvplot.bar(
    xlabel='Year',
    ylabel='Housing Units',
    title='Housing Units in San Francisco from 2010 to 2016'
).opts(yformatter="%.0f")
```

# Compare the Average Sale Prices by Neighborhood

```python
# Create a new DataFrame that groups the original DataFrame by year and neighborhood. Aggregate the results by the mean of the groups.
average_prices_by_neighborhood = sfo_data.groupby(['year', 'neighborhood']).mean()
display(average_prices_by_neighborhood)
```

```python
# Filter out the “housing_units” column to create a DataFrame that includes only the sale_price_sqr_foot and gross_rent averages per year.
average_prices_by_neighborhood = average_prices_by_neighborhood.drop(columns = ['housing_units'])
display(average_prices_by_neighborhood)
```

```python
# Create an interactive line plot with hvPlot that visualizes both sale_price_sqr_foot and gross_rent. Set the x-axis parameter to the year (x="year"). Use the groupby parameter to create an interactive widget for neighborhood.
average_prices_by_neighborhood.hvplot.line(
    x="year",
    y=["sale_price_sqr_foot", "gross_rent"],
    xlabel='Year',
    ylabel='Price per Square Foot',
    groupby='neighborhood',
    title='Average Price per Square Foot and Gross Rent in San Francisco from 2010 to 2016'
).opts(yformatter="%.0f")
```

## Build an Interactive Neighborhood Map

```python
# Read the neighborhood_coordinates.csv file from the Resources folder into the notebook, and create a DataFrame named neighborhood_locations_df. 
#Be sure to set the index_col of the DataFrame as “Neighborhood”.
neighborhood_data = pd.read_csv(Path('/Users/TFitz/Documents/School/FINTECH-Homework-Modules/Module 6 Challenge/Starter_Code/Resources/neighborhoods_coordinates.csv'))
neighborhood_data.set_index('Neighborhood', inplace = True)
display(neighborhood_data)
```

```python
# Using the original sfo_data_df Dataframe, create a DataFrame named all_neighborhood_info_df that groups the data by neighborhood. 
#Aggregate the results by the mean of the group.
all_neighborhood_info_df = sfo_data.groupby('neighborhood').mean()
display(all_neighborhood_info_df)
```

```python
# Review the two code cells that concatenate the neighborhood_locations_df DataFrame with the all_neighborhood_info_df DataFrame.
all_neighborhoods_df = pd.concat([neighborhood_data, all_neighborhood_info_df], axis = 1, join = 'inner')
all_neighborhoods_df.reset_index(inplace = True)
all_neighborhoods_df.rename(columns = {'index': 'Neighborhood'}, inplace = True)
display(all_neighborhoods_df)
```

```python
#Using hvPlot with GeoViews enabled, create a points plot for the all_neighborhoods_df DataFrame.
points_plot = all_neighborhoods_df.hvplot.points(
    hover_cols=['Neighborhood'],
    x='Lon',
    y='Lat',
    geo=True,
    size='sale_price_sqr_foot',
    frame_width=700,
    frame_height=500,
    color='gross_rent',
    title='Average Sale Price Per Square Foot and Gross Rent in San Francisco'
).opts(yformatter="%.0f")
points_plot
```

Use the interactive map to answer the following question:

Which neighborhood has the highest gross rent, and which has the highest sale price per square foot?

Answer: Westwood park has the highest gross rent, and Union Square District has the highest price per square foot.
